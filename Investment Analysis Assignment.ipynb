```python
#checklist 1: Cleaning Data
# Read the comapnies data
# companies3 is same data set provided in the upgrad portal just the name is renamed.

companies = pd.read_csv('companies3.txt',sep="\t",encoding='latin1')
```

```python
companies
```

```python
#Analysing the company data.
companies.isnull().sum()
```

```python
#removing special characters  from companies data frame
companies['permalink']=companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
```

```python
companies['name']= companies.name.str.encode('utf-8').str.decode('ascii','ignore')
```

```python
#removing special characters from rounds2
rounds2['company_permalink']=rounds2[['company_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii','ignore'))
```

```python
rounds2['funding_round_permalink']=rounds2[['funding_round_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii','ignore'))
```

```python
companies['name']= companies['name'].str.lower()
```

```python
# 1.Extract the primary sector of each category list from the category_list column, 
# first string before vertical bar will be considered primary sector

primary_sector = master_frame_venture['category_list'].str.split('|', n=1,expand=True)
```

```python
# Q2. Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors 
#(Note that ‘Others’ is also considered one of the main sectors)

master_frame_venture['Primary_Sector'] = primary_sector[0]
```

```python
mapping['category_list'] = mapping['category_list'].str.replace('0','na')
```

```python
mapping.columns
```

```python
#  using 'category_list' as the column in front of which we need to get the columns containing 1.

#pandas.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', col_level=None)

melted_df = pd.melt(mapping, id_vars=['category_list'],value_vars=['Automotive & Sports', 'Blanks',
       'Cleantech / Semiconductors', 'Entertainment', 'Health',
       'Manufacturing', 'News, Search and Messaging', 'Others',
       'Social, Finance, Analytics, Advertising'])
```

```python
melted_df
```

```python
# extracting those columns which has 1 in the respective cell

melted_df = melted_df[melted_df.value == 1]
```

```python
melted_df
```

```python
# changing the name of the column 'category_list' to 'Primary Sector' and vairable to 'main_sector'

melted_df = melted_df.rename(columns={'category_list':'Primary_Sector', 'variable':'main_sector'})
```

```python
melted_df
```

```python
# creating a merged data frame where each primary sector mapped to its main sector

master_dataframe = pd.merge(master_frame_venture,melted_df,on='Primary_Sector',how='inner')
```

```python
# checklist 5 : Sector Analysis 2

# creating a data frame for which the investment is between 5 to 15 million 

Spark_Funds_5M_15M = master_dataframe[master_dataframe['raised_amount_usd'].between(5000000,15000000,inclusive=True)]
```

```python
Spark_Funds_5M_15M
```

```python
#Now, the aim is to find out the most heavily invested main sectors in each of the three countries 
#(for funding type FT and investments range of 5-15 M USD).
#Create three separate data frames D1, D2 and D3 for each of the three countries containing the 
#observations of funding type FT falling within the 5-15 million USD range.

#All the columns of the master_frame along with the primary sector and the main sector

#The total number (or count) of investments for each main sector in a separate column

#The total amount invested in each main sector in a separate column

# According to problem statement creating 3 dataframes D1,D2,D3


D1 = Spark_Funds_5M_15M[Spark_Funds_5M_15M['country_code']=='USA']
D1 = D1.assign(count_of_invest = D1.groupby('main_sector')['main_sector'].transform('count'))
D1 = D1.assign(sum_of_invest =D1.groupby('main_sector')['raised_amount_usd'].transform('sum'))
```
